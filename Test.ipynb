# Automated analysis of whole-cell calcium signals

## 1 - Load some libraries

```python
%pylab inline
from __future__ import division
import scipy.signal as signal
import numpy
import matplotlib
import os
import scipy
from matplotlib import pylab, mlab, pyplot
from scipy.stats import gaussian_kde
np = numpy
plt = pyplot
import random
from scipy import ndimage
```

## 3 - Baseline normalise the signals

### Raw signals will be baseline-corrected by dividing the intensity at each time point by the average value of a baseline period. These signals are F/F0.  For this dataset, we will specifcy a 100-frame long baseline period starting at frame 0 and ending at frame 99.

```python
b0, b1 = 0,99
```

### Now normalise and plot the signals

```python
baselined_signals = [sig / numpy.average(sig[b0:b1]) for sig in raw_signals]

pylab.figure(figsize=(8,5))
pylab.xlabel('Frame no.')
pylab.ylabel('F/F0')
for sig in baselined_signals:
    pylab.plot(sig)
```

### As you can see, the baseline (first 100 frames) of each trace is now equal to 1 and all the traces are layered on top of each other. Lets try some alternative ways of displaying the data:

#### First, lets colour all traces black, make all signals transparent, and plot the average value over the top in red.

```python
for i,sig in enumerate(baselined_signals):
            #print i,
            pylab.plot(sig, alpha=0.05,linewidth=1, color='k')
pylab.plot(numpy.average(baselined_signals, axis=0), alpha=1,linewidth=2, color='r')
```

#### We can also try offsetting each of the signals along the y-axis.

```python
pylab.figure(figsize=(8,20))
pylab.xlabel('Frame no.')
pylab.ylabel('F/F0')
for i,sig in enumerate(baselined_signals):
    if sum(sig)!=0:
        pylab.plot(sig+i/5,linewidth=1,color='k')
#pylab.ylim((0.9,1.1))
```

### Because there are so many signals in this dataset, we are going to extract just a single trace and demonstrate how to identify peaks in the signal, and extract information regarding the time at which these occur and the amplitude

### Lets just use a portion of a single trace:

```python
sig_portion = baselined_signals[8][0:599]
pylab.plot(sig_portion,color='k', linewidth=3)
```

### 4 - Smooth (i.e. low pass filter) the baseline normalised signals

#### This is done to remove any high-frequency noise in the signal that may hamper any efforts to identify peaks in the signal

```python
smoothed_sig_portion = signal.savgol_filter(sig_portion, 21, 3)
```

### Lets plot the original and the smoothed signal on the same plot to check that we havenot went overboard

```python
pylab.plot(sig_portion, color="k",lw=5, label="Original")
pylab.plot(smoothed_sig_portion, color="red", lw=5, label="Smoothed")
pylab.legend(loc="best")
```

## 05 - Let us now differentiate the function with respect to time

### First, lets define a function that differentiates by convolving with the derivative of a guassian kernal

```python
def diff(sig, n):
    dx = 1/n
    ddt = ndimage.gaussian_filter1d(sig, sigma=5, order=1, mode='nearest') / dx
    ddt = numpy.array(ddt)
    return ddt
```

```python
freq=10 # This is the acquisition frequency
ddt_sig_portion = diff(smoothed_sig_portion, freq)

# --------------------------------------
# PLOT: Smooth Baselined signals
# --------------------------------------
pylab.figure(figsize=(8,5))
pylab.title('Derivate Signal', fontsize=20)
pylab.xlabel('Frame no.')
pylab.ylabel('d(F/F0)/dt')
pylab.plot(ddt_sig_portion, color="k",lw=5, label="Original")
```

### Lets find all of the x-values (frame no) where this derivate signal crosses the x-axis. These are called "zero-crossings"

```python
zero_crossings = numpy.where(numpy.diff(numpy.sign(ddt_sig_portion)))[0]
print zero_crossings
```

### Let us show all of the zero crossings on a plot:

```python
pylab.plot(ddt_sig_portion,color='k', linewidth=3)
for zero_crossing in zero_crossings:
    plt.axvline(zero_crossing, color='r')
```

### As you can see, between each consecutive set of zero-crossings, there is a peak or nadir in the signal. However, some of these are just because of baseline noise. We are interested in finding out which ones actually correspond to peaks

```python
#Calculate the periods - the number of frames in between each consecutive pair of zero-crossings.
period_lengths = numpy.diff(zero_crossings)
print period_lengths
```

### Using these periods, we can chop the derivate signal up into portions...

```python
x_axis = range(len(ddt_sig_portion)) #Set up the x axis
bins = [ddt_sig_portion[indice:indice+diff] for indice, diff in zip(zero_crossings, period_lengths)] #Divide the signal bins
```

### ... and then decide if the portion contains a peak or a nadir...

```python
peaks = []
nadirs = []
for idx in range(1,len(bins)):               #Set up a loop to go through each of the bin
    if bins[idx].max() > abs(bins[idx].min()): #Check if the critical point is s maxima
        peaks.append(bins[idx].max())
    else:                                      #If it is not a maxima, then it must be a minima
        nadirs.append(bins[idx].min())
print "We found this many nadirs: ",len(nadirs)
print "We found this many peaks: ",len(peaks)
```

### ... and extract the x-value at whuch each of these peaks/nadir occur:

```python
#GET THE INDICE (i.e. frame numbers) for each of the peaks VALUE OF THE PEAKS
peaks_idx = [x_axis[numpy.where(ddt_sig_portion==peak)[0][0]] for peak in peaks]
nadirs_idx = [x_axis[numpy.where(ddt_sig_portion==peak)[0][0]] for nadir in nadirs]
print peaks_idx
print nadirs_idx
```

### Let us show the locations of all of the identified peaks on a plot (we do not really care about the nadirs):

```python
pylab.plot(ddt_sig_portion,color='k', linewidth=3)
for peak in peaks_idx:
    plt.axvline(peak, color='r')
#plt.xlim(0,500)
```

### Let us now get the xy valyes of each identified peak and filter out those below a threshold level (i.e. baseline noise)

```python
#List of positive peaks
peak_values = [(x,y) for x,y in zip(peaks_idx,peaks)]

print "Coordinates of all peaks = "
print peak_values

stdev = numpy.std(ddt_sig_portion[0:99]) # Calculate the standard deviation of the the baseline period.
thresh_peak_values = [item for item in peak_values if item[1]>3*stdev] #Threshold thm

print "Coordinates of all peaks above threshold = "
print thresh_peak_values
```

```python
pylab.plot(ddt_sig_portion,color='k', linewidth=3)

for peak in thresh_peak_values:
    plt.scatter(peak[0],peak[1], s=200, color='r')
    plt.axvline(peak[0], color='r')
plt.xlim(0,599)
```

### As you can see, we now have the x and y values for each peak in the derivate signal.

### However, these peaks correspond to where the original signal is experiencing its maximum rate of change. This is not the peak of the original F/F0. Lets show it on a graph:

```python
pylab.plot(sig_portion,color='k', linewidth=3)
for peak in thresh_peak_values:
    plt.axvline(peak[0], color='r')
plt.xlim(0,599)
```

### This means that we cannot simply readout the vales of the F/F0 signal at these x-axis locations. We need to find the peak values!

```python
ddt_peak_x = [i[0] for i in thresh_peak_values] #x-values of peaks in ddt signal
ddt_peak_y = [i[1] for i in thresh_peak_values] #y-values of peaks in ddt signal
```

```python
sig_peak_y = [numpy.max(sig_portion[i:i+30]) for i in ddt_peak_x] #x-values of peaks in ddt signal
sig_peak_x = [numpy.where(sig_portion==peak)[0][0] for peak in sig_peak_y] #y-values of peaks in ddt signal
```

```python
pylab.plot(sig_portion,color='k', linewidth=3)
for peak in zip(sig_peak_x,sig_peak_y):
    plt.scatter(peak[0],peak[1], s=200, color='r')
    plt.axvline(peak[0], color='r')
plt.xlim(0,599)
```

### As you can see, we now have the time at which all the peaks occur. We also have the amplitude of each peak. Using this data we can calculate many signalling metrics for this signal. These include: the magnitude of the first peak, the frequency of oscillation, the number of peaks occuring throughout the length of the signal (i.e. oscillation freuency). Even better, by applying this code in a loop, we can work out these values for hundreds of signals without much work at all.
